## Estructura de almacenamiento de archivos
 * Organizados en carpetas
 * Con nombres estandarizados
 * Para que el programa funcione el intervalo de meses debe estar completo, o sea no puede faltar un mes dentro de cierto intervalo
 
 



```python
import pandas as pd
import seaborn as sns
import os
```

```python
# Generacion listado de nombres estandarizados
a1=['ST','VS']
a2=['057','A14','A15','A16','1604']
a3=['PC','PS','NC','NS']
# Generacion del listado de nombres
narch=[]
for j in a2:
    for i in a1:
        for k in range(len(a3)):
            g='0'+str(k+1)+'_' + i + '_' + j + '_' + a3[k]+'.csv'
            narch.append(g)      
```

```python

def e140(carp,MESAR):

        carpeta=["01_INSUMO/"+ carp + "/SICA ST "+MESAR+" 2020/01_ST_E.14.0/","01_INSUMO/"+ carp + "/SICA VS "+MESAR+" 2020/01_VS_E.14.0/"]
        cont=0
        numeral=0                           # numeral de la carpeta 
        for h in range(2):
            print('\n'+carpeta[h])               # Verificacion de la carpeta
            for j in range(numeral,numeral+4):
                print(narch[j])             # Verificacion de archivos

                car=['Servicio','Indicador','C','Fecha','Tipo Regimen','F','Periodo','H']
                tr=pd.read_csv(carpeta[h]+narch[j], header=0)       #t=pd.read_csv(narch[p], header=None, names=cols)
                y=list(tr.columns)
                
                
                tr=pd.read_csv(carpeta[h]+narch[j], header=1, thousands='.' , decimal= ',',  encoding='utf-8')      #t=pd.read_csv(narch[p], header=None, names=cols)
                 
                
                
#                tr.columns=list(tr.loc[1])
                tr.drop([0, 1],axis=0, inplace=True)
                tr.dropna(axis=1,how='all',inplace=True)
                tr.drop(tr[tr['Resultado'].isnull()].index, inplace = True)    #elimina las filas que tengan valores nulos en resultado
                for k in range(len(car)):                                      #asignacion de campos complementarios que describen la base de datos 
                    tr[car[k]]=y[k]
                tr['Fuente']=narch[j]  
                cont+=1   
                if (cont==1):
                    tr057=tr
                    t057=tr
                else:
                    tr057=pd.concat([tr057,tr])
                    
            numeral=numeral+4                                    #Especificacion del rango de archivos en la carpeta  
        tr057['MMES']=MESAR   
        tr057.to_csv('02_CONSOLIDADO/057 CON/057_'+MESAR+'.csv', decimal=',')    #  ) #, thousands='.') #,         #El resultado se almacena como csv 
```

```python
def eA14(carp,MESAR):
        carpeta=["01_INSUMO/"+ carp + "/SICA ST "+MESAR+" 2020/02_ST_A14/","01_INSUMO/"+ carp + "/SICA VS "+MESAR+" 2020/02_VS_A14/"]
        cont=0
        numeral=8                                  # Numeral de la carpeta 
        for h in range(2):                         # ESTAS PRUEBAS SE PUEDEN HACER CON 1 O  2
            print(carpeta[h])                      # Verificacion de la carpeta
            for j in range(numeral,numeral+4):
                print(narch[j])                    # Verificacion de archivos
                car=['Servicio','Indicador','C','Fecha','Tipo Regimen','F','Periodo','H']
                tr=pd.read_csv(carpeta[h]+narch[j], header=None)  
                y=list(tr.loc[0])
                tr.columns=list(tr.loc[1])
                tr.drop([0, 1],axis=0, inplace=True)
                tr.dropna(axis=1,how='all',inplace=True)
                tr.drop(tr[tr['Resultado'].isnull()].index, inplace = True)    #elimina las filas que tengan valores nulos en resultado
                for k in range(len(car)):                                      #asignacion de campos complementarios que describen la base de datos 
                    tr[car[k]]=y[k]
                tr['Fuente']=narch[j]  
                cont+=1   
                if (cont==1):
                    trA14=tr
                else:
                    trA14=pd.concat([trA14,tr])
                    
                    
            numeral=numeral+4                                    #Especificacion del rango de archivos en la carpeta  
        set(trA14['Fuente'])
        set(trA14['Indicador'])
        
        trA14['MMES']=MESAR 
        trA14.to_csv('02_CONSOLIDADO/A14 CON/A14_' + MESAR+'.csv', decimal=',')
```
